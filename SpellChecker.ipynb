```python
import torch
from transformers import BertTokenizer, BertForMaskedLM
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import words
from spellchecker import SpellChecker

# Download NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('words')

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')
model.eval()

# Set of English words for basic spell checking
english_vocab = set(words.words())

# Initialize SpellChecker
spell = SpellChecker()

def tokenize_sentence(sentence):
    tokens = word_tokenize(sentence.lower())
    return tokens

def correct_spelling(tokens):
    corrected_tokens = []

    for token in tokens:
        if token.lower() not in english_vocab:
            # Use BERT for contextual prediction
            masked_index = tokens.index(token)
            tokens[masked_index] = '[MASK]'

            # Convert tokens to IDs
            token_ids = tokenizer.convert_tokens_to_ids(tokens)
            token_tensor = torch.tensor([token_ids])

            # Predict all tokens
            with torch.no_grad():
                outputs = model(token_tensor)
                predictions = outputs[0]

            predicted_index = torch.argmax(predictions[0, masked_index]).item()
            predicted_token = tokenizer.convert_ids_to_tokens([predicted_index])[0]
            corrected_tokens.append(predicted_token)
        else:
            corrected_tokens.append(token)

    return corrected_tokens

def spell_check_and_correct(sentence):
    tokens = tokenize_sentence(sentence)
    corrected_tokens = []
    misspelled_words = []

    for token in tokens:
        corrected_word = spell.correction(token)
        if corrected_word is None:
            corrected_tokens.append(token)  # Keep original token if no correction found
        else:
            corrected_tokens.append(corrected_word)
            if token != corrected_word:
                misspelled_words.append(token)

    # Construct the corrected sentence
    corrected_sentence = ' '.join(corrected_tokens)

    # Remove trailing punctuation marks (.,;:"' etc.)
    corrected_sentence = corrected_sentence.strip(".,;:\"'")

    return misspelled_words, corrected_sentence

while True:
    user_input = input("Enter a sentence (or 'q' to quit): ")

    if user_input.lower() == 'q':
        break

    misspelled_words, corrected_sentence = spell_check_and_correct(user_input)

    if misspelled_words:
        print(f"Original: {user_input}")
        print(f"Misspelled words: {misspelled_words}")
        print(f"Corrected: {corrected_sentence}\n")
    else:
        print(f"Original: {user_input}")
        print(f"No misspelled words found. Corrected: {corrected_sentence}\n")
```
